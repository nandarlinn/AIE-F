```python
from pyngrok import ngrok
import subprocess, threading, time

# Start Streamlit in background
def run():
    subprocess.run(["streamlit", "run", "app_cleaned.py",
                    "--server.port", "8501",
                    "--server.headless", "true"])

thread = threading.Thread(target=run)
thread.daemon = True
thread.start()

time.sleep(10)  # wait for streamlit to start, increased to 10 seconds
ngrok.set_auth_token("add your token here")

# Kill any existing ngrok tunnels to free up the port
ngrok.kill()

# Create public URL
public_url = ngrok.connect(8501)
print("✅ Your chatbot is live at:", public_url)
```

## Setup

```python
from pathlib import Path
```

```python
ROOT = Path("..").resolve()

DATA_DIR = ROOT / "data"

ORIGINAL_DIR = DATA_DIR / "g2p-par"
CLEANED_DIR = DATA_DIR / "cleaned"
LOG_DIR = DATA_DIR / "logs"

SYL_NORM = ROOT / "syl-normalizer" / "syl_normalizer.py"
SYL_DICT = ROOT / "syl-normalizer" / "final_syl_dictionary_13Feb2024.sorted.txt"
```

## Preprocess Data
### Normalize Burmese Syllables in `.my` Files

`.ph` files do not need normalization because they contain English phonemes representing how Burmese words sound.

```python
!python {SYL_NORM} \
    --dictionary {SYL_DICT} \
    --frequency 2 \
    --input {ORIGINAL_DIR}/train.my \
    --output {CLEANED_DIR}/train.my \
    --log {LOG_DIR}/cleaning_train_my.log \
    --error-output {LOG_DIR}/cleaning_train_my.errors \
    --fuzzy-distance 0
```

```python
!wc {ORIGINAL_DIR}/train.my
!wc {CLEANED_DIR}/train.my
```

```python
!python {SYL_NORM} \
    --dictionary {SYL_DICT} \
    --frequency 2 \
    --input {ORIGINAL_DIR}/dev.my \
    --output {CLEANED_DIR}/dev.my \
    --log {LOG_DIR}/cleaning_dev_my.log \
    --error-output {LOG_DIR}/cleaning_dev_my.errors \
    --fuzzy-distance 0
```

```python
!wc {ORIGINAL_DIR}/train.my
!wc {CLEANED_DIR}/train.my
```

```python
!python {SYL_NORM} \
    --dictionary {SYL_DICT} \
    --frequency 2 \
    --input {ORIGINAL_DIR}/test.my \
    --output {CLEANED_DIR}/test.my \
    --log {LOG_DIR}/cleaning_test_my.log \
    --error-output {LOG_DIR}/cleaning_test_my.errors \
    --fuzzy-distance 0
```

```python
!wc {ORIGINAL_DIR}/test.my
!wc {CLEANED_DIR}/test.my
```

# 00 Exact Template Diagnostic Compare

Final exact-template PBSMT notebook.

## Purpose

Your original `pbsmt-big-normalize` result is much better, so this notebook checks the real differences instead of guessing.

It compares:

- original config: `pbsmt-big-normalize/baseline/my-ph/config.baseline.my-ph`
- previous final generated config: `pbsmt_final_clean_experiments/5g_phrase9_original_search_MAIN/my-ph/config.baseline.my-ph`
- original output/reference vs generated output/reference

The goal is to prove whether the issue is config, data/SGM, or output generation.

```python
from pathlib import Path
import subprocess, re, shutil, difflib, os

BASE_DIR = Path('/media/kyalkalay/Data/AI-Class/assignment-6_aung_hein')
G2P_DIR = BASE_DIR / 'g2p-par'
MOSES_SRC_DIR = Path('/home/kyalkalay/tools/mosesdecoder')
MOSES_SCRIPT_DIR = MOSES_SRC_DIR / 'scripts'
EXPERIMENT_PERL = MOSES_SCRIPT_DIR / 'ems' / 'experiment.perl'

# IMPORTANT: use the original high-BLEU template exactly.
ORIGINAL_EXP_DIR = BASE_DIR / 'pbsmt-big-normalize'
ORIGINAL_TEMPLATE = ORIGINAL_EXP_DIR / 'config.baseline'
ORIGINAL_BASELINE_DIR = ORIGINAL_EXP_DIR / 'baseline'

FINAL_EXACT_DIR = BASE_DIR / 'pbsmt_final_exact_template_experiments'

print('BASE_DIR:', BASE_DIR)
print('Original template:', ORIGINAL_TEMPLATE, ORIGINAL_TEMPLATE.exists())
print('Original baseline:', ORIGINAL_BASELINE_DIR, ORIGINAL_BASELINE_DIR.exists())
print('Final exact dir:', FINAL_EXACT_DIR)
```

```python
orig_cfg = ORIGINAL_BASELINE_DIR / 'my-ph' / 'config.baseline.my-ph'
bad_cfg = BASE_DIR / 'pbsmt_final_clean_experiments' / '5g_phrase9_original_search_MAIN' / 'my-ph' / 'config.baseline.my-ph'

print('orig_cfg exists:', orig_cfg.exists(), orig_cfg)
print('bad_cfg exists:', bad_cfg.exists(), bad_cfg)

if orig_cfg.exists() and bad_cfg.exists():
    a = orig_cfg.read_text(encoding='utf-8', errors='ignore').splitlines()
    b = bad_cfg.read_text(encoding='utf-8', errors='ignore').splitlines()
    # Hide expected path differences to make real setting differences visible.
    def normalize(lines):
        out = []
        for x in lines:
            x = re.sub(r'working-dir\s*=.*', 'working-dir = <WORKDIR>', x)
            out.append(x)
        return out
    diff = list(difflib.unified_diff(normalize(a), normalize(b), fromfile='original', tofile='generated', lineterm=''))
    print('\n'.join(diff[:400]))
    print('\nTotal diff lines:', len(diff))
```

```python
def read_lines(p):
    return p.read_text(encoding='utf-8', errors='ignore').splitlines() if p.exists() else []

pairs = [
    ('input', ORIGINAL_BASELINE_DIR/'my-ph/evaluation/test.input.txt.1', BASE_DIR/'pbsmt_final_clean_experiments/5g_phrase9_original_search_MAIN/my-ph/evaluation/test.input.txt.1'),
    ('reference', ORIGINAL_BASELINE_DIR/'my-ph/evaluation/test.reference.txt.1', BASE_DIR/'pbsmt_final_clean_experiments/5g_phrase9_original_search_MAIN/my-ph/evaluation/test.reference.txt.1'),
    ('output', ORIGINAL_BASELINE_DIR/'my-ph/evaluation/test.output.1', BASE_DIR/'pbsmt_final_clean_experiments/5g_phrase9_original_search_MAIN/my-ph/evaluation/test.output.1'),
]

for name, a, b in pairs:
    la, lb = read_lines(a), read_lines(b)
    print('\n##', name)
    print('A:', a, 'lines=', len(la), 'exists=', a.exists())
    print('B:', b, 'lines=', len(lb), 'exists=', b.exists())
    if la and lb:
        same = sum(x == y for x, y in zip(la, lb))
        print('same prefix-aligned lines:', same, '/', min(len(la), len(lb)))
        for i, (x, y) in enumerate(zip(la, lb), start=1):
            if x != y:
                print('first diff line:', i)
                print('original:', x[:300])
                print('generated:', y[:300])
                break
```

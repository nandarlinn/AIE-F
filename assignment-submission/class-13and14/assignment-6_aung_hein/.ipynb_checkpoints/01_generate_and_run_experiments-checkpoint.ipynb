## Purpose

Generate final clean configs for 3-gram, 5-gram, and 7-gram PBSMT experiments.

Final design choice:

```text
No forced -distortion-limit 0
```

This matches the original successful 5-gram setting more closely.

````python
from pathlib import Path
import subprocess, re, os, json, html, shutil

# === Final project paths ===
BASE_DIR = Path('/media/kyalkalay/Data/AI-Class/assignment-6_aung_hein')
G2P_DIR = BASE_DIR / 'g2p-par'
SGM_DIR = G2P_DIR / 'sgm'
MOSES_SRC_DIR = Path('/home/kyalkalay/tools/mosesdecoder')
MOSES_SCRIPT_DIR = MOSES_SRC_DIR / 'scripts'
MOSES_BIN_DIR = MOSES_SRC_DIR / 'bin'
EXPERIMENT_PERL = MOSES_SCRIPT_DIR / 'ems' / 'experiment.perl'
EXTERNAL_BIN_DIR = Path('/usr/local/bin')

print('BASE_DIR:', BASE_DIR)
print('G2P_DIR:', G2P_DIR)
print('Moses:', MOSES_SRC_DIR)
print('EMS:', EXPERIMENT_PERL)
````

````python
def sh(cmd, cwd=None, check=True):
    """Run a shell command and stream output."""
    print(f"$ {cmd}")
    return subprocess.run(cmd, shell=True, cwd=str(cwd) if cwd else None, text=True, check=check)


def make_moses_ems_config(
    working_dir,
    input_ext,
    output_ext,
    order=5,
    max_phrase_length=9,
    lm_prune='0 0 1',
    lm_memory='20%',
    lexicalized_reordering='msd-bidirectional-fe',
    eval_decoder_settings='-search-algorithm 1 -cube-pruning-pop-limit 5000 -s 5000',
    tuning_decoder_settings='',
    nbest=100,
    cores=None,
):
    """Create final Moses EMS config.

    Final important rule:
    - Do NOT force `-distortion-limit 0`.
    - Tuning decoder-settings should remain empty unless an experiment explicitly tests search only.
    - Evaluation decoder settings follow the original assignment config.
    """
    cores_line = f"cores = {cores}\n" if cores else ""
    return f"""
[GENERAL]
working-dir = {working_dir}
input-extension = {input_ext}
output-extension = {output_ext}
pair-extension = {input_ext}-{output_ext}

### directories that contain tools and data
moses-src-dir = {MOSES_SRC_DIR}
moses-bin-dir = $moses-src-dir/bin
moses-script-dir = $moses-src-dir/scripts
external-bin-dir = {EXTERNAL_BIN_DIR}
myph-data = {G2P_DIR}

### basic tools
decoder = $moses-bin-dir/moses
generic-parallelizer = $moses-script-dir/ems/support/generic-parallelizer.perl
{cores_line}
[CORPUS]
max-sentence-length = 80

[CORPUS:myph]
raw-stem = $myph-data/train_clean

[LM]
lmplz = $moses-bin-dir/lmplz
lm-training = "$moses-script-dir/ems/support/lmplz-wrapper.perl -bin $moses-bin-dir/lmplz"
settings = "--prune '{lm_prune}' -T $working-dir/lm -S {lm_memory} --discount_fallback"
order = {order}
lm-binarizer = $moses-bin-dir/build_binary
type = 8

[LM:myph]
raw-corpus = $myph-data/train_clean.$output-extension

[INTERPOLATED-LM]
lm-binarizer = $moses-bin-dir/build_binary
type = 8

[MML] IGNORE

[TRAINING]
script = $moses-script-dir/training/train-model.perl
parallel = yes
run-giza-in-parts = 5
alignment-symmetrization-method = grow-diag-final-and
lexicalized-reordering = {lexicalized_reordering}
max-phrase-length = {max_phrase_length}
score-settings = "--GoodTuring"

[TUNING]
tuning-script = $moses-script-dir/training/mert-moses.pl
tuning-settings = "-mertdir $moses-bin-dir"
raw-input = $myph-data/dev_clean.$input-extension
raw-reference = $myph-data/dev_clean.$output-extension
nbest = {nbest}
filter-settings = ""
decoder-settings = "{tuning_decoder_settings}"

[EVALUATION]
decoder-settings = "{eval_decoder_settings}"
wrapping-script = "$moses-script-dir/ems/support/wrap-xml.perl $output-extension"
nist-bleu = $moses-script-dir/generic/mteval-v13a.pl
nist-bleu-c = "$moses-script-dir/generic/mteval-v13a.pl -c"
analysis = $moses-script-dir/ems/support/analysis.perl
analyze-coverage = yes
report-segmentation = yes
report-precision-by-coverage = yes

[EVALUATION:test]
input-sgm = $myph-data/sgm/test.$input-extension.src.sgm
reference-sgm = $myph-data/sgm/test.$output-extension.ref.sgm
wrapping-frame = $input-sgm

[REPORTING]
""".strip() + "\n"


def write_experiment_configs(exp_name, variants, directions=(('my','ph'), ('ph','my')), clean_existing=False):
    """Write clean config files into BASE_DIR/exp_name/<variant>/<pair>/.

    If clean_existing=True, the whole experiment folder is deleted first.
    Use with care.
    """
    exp_root = BASE_DIR / exp_name
    if clean_existing and exp_root.exists():
        shutil.rmtree(exp_root)
    exp_root.mkdir(parents=True, exist_ok=True)

    written = []
    for variant in variants:
        for src, trg in directions:
            pair = f'{src}-{trg}'
            work = exp_root / variant['name'] / pair
            work.mkdir(parents=True, exist_ok=True)
            cfg = make_moses_ems_config(
                working_dir=work,
                input_ext=src,
                output_ext=trg,
                order=variant['order'],
                max_phrase_length=variant.get('max_phrase_length', 9),
                lm_prune=variant.get('lm_prune', '0 0 1'),
                lm_memory=variant.get('lm_memory', '20%'),
                lexicalized_reordering=variant.get('lexicalized_reordering', 'msd-bidirectional-fe'),
                eval_decoder_settings=variant.get('eval_decoder_settings', '-search-algorithm 1 -cube-pruning-pop-limit 5000 -s 5000'),
                tuning_decoder_settings=variant.get('tuning_decoder_settings', ''),
                nbest=variant.get('nbest', 100),
                cores=variant.get('cores'),
            )
            cfg_path = work / f'config.baseline.{pair}'
            cfg_path.write_text(cfg, encoding='utf-8')
            written.append(cfg_path)
    print(f'Wrote {len(written)} config files under {exp_root}')
    return written


def run_configs(config_paths, no_graph=True):
    """Run Moses EMS. This can take a long time."""
    for cfg in config_paths:
        log_path = cfg.parent / 'run.log'
        cmd = f"perl {EXPERIMENT_PERL} -config {cfg} -exec {'-no-graph' if no_graph else ''} 2>&1 | tee {log_path}"
        sh(cmd, cwd=cfg.parent, check=False)
````

## Final experiment variants

````python
FINAL_VARIANTS = [
    # Lower-context baselines
    {'name': '3g_phrase5_original_search', 'order': 3, 'max_phrase_length': 5, 'lm_prune': '0 0 1', 'nbest': 100},
    {'name': '3g_phrase7_original_search', 'order': 3, 'max_phrase_length': 7, 'lm_prune': '0 0 1', 'nbest': 100},
    {'name': '3g_phrase5_less_pruned', 'order': 3, 'max_phrase_length': 5, 'lm_prune': '0 0 0', 'nbest': 100},

    # Main experiment: original-like 5g setup
    {'name': '5g_phrase9_original_search_MAIN', 'order': 5, 'max_phrase_length': 9, 'lm_prune': '0 0 1', 'nbest': 100},
    {'name': '5g_phrase9_less_pruned', 'order': 5, 'max_phrase_length': 9, 'lm_prune': '0 0 0', 'nbest': 100},
    {'name': '5g_phrase11_original_search', 'order': 5, 'max_phrase_length': 11, 'lm_prune': '0 0 1', 'nbest': 100},

    # Higher-context tests
    {'name': '7g_phrase9_original_search', 'order': 7, 'max_phrase_length': 9, 'lm_prune': '0 0 1', 'nbest': 100},
    {'name': '7g_phrase11_original_search', 'order': 7, 'max_phrase_length': 11, 'lm_prune': '0 0 1', 'nbest': 100},
    {'name': '7g_phrase9_less_pruned', 'order': 7, 'max_phrase_length': 9, 'lm_prune': '0 0 0', 'nbest': 100},
]

configs = write_experiment_configs(
    exp_name='pbsmt_final_clean_experiments',
    variants=FINAL_VARIANTS,
    directions=(('my','ph'), ('ph','my')),
    clean_existing=False,
)
for c in configs:
    print(c)
````

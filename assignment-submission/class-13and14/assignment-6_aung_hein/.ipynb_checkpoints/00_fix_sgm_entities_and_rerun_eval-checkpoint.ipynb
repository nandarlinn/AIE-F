# 00 Fix SGM Entities and Rerun Evaluation

Fix SGM XML escaping and rerun evaluation.

## Root cause

The model outputs are mostly similar, but the generated reference file contains HTML/XML entity text such as:

```text
te&#x27; te&#x27; pjaun
```

or after another escaping pass:

```text
te&amp;#x27; te&amp;#x27; pjaun
```

The original reference is:

```text
te' te' pjaun
```

This corrupts many phoneme references and drops BLEU even when the model output is good.

This notebook regenerates SGM files using XML escaping that **does not escape apostrophes** and also unescapes any existing entity text before writing.

````python
from pathlib import Path
import html, re, subprocess, shutil, os

BASE_DIR = Path('/media/kyalkalay/Data/AI-Class/assignment-6_aung_hein')
G2P_DIR = BASE_DIR / 'g2p-par'
SGM_DIR = G2P_DIR / 'sgm'
MOSES_SRC_DIR = Path('/home/kyalkalay/tools/mosesdecoder')
MOSES_SCRIPT_DIR = MOSES_SRC_DIR / 'scripts'
EXPERIMENT_PERL = MOSES_SCRIPT_DIR / 'ems' / 'experiment.perl'
MTEVAL = MOSES_SCRIPT_DIR / 'generic' / 'mteval-v13a.pl'

ORIG_EVAL = BASE_DIR / 'pbsmt-big-normalize' / 'baseline' / 'my-ph' / 'evaluation'
FINAL_EXACT = BASE_DIR / 'pbsmt_final_exact_template_experiments' / '5g_original_template_MAIN' / 'my-ph'
FINAL_EXACT_EVAL = FINAL_EXACT / 'evaluation'
FINAL_CLEAN = BASE_DIR / 'pbsmt_final_clean_experiments' / '5g_phrase9_original_search_MAIN' / 'my-ph'
FINAL_CLEAN_EVAL = FINAL_CLEAN / 'evaluation'

print('BASE_DIR:', BASE_DIR)
print('SGM_DIR:', SGM_DIR)
print('MTEVAL:', MTEVAL, MTEVAL.exists())
print('FINAL_EXACT_EVAL:', FINAL_EXACT_EVAL, FINAL_EXACT_EVAL.exists())
````

## 1. Compare original vs generated reference corruption

````python
def read_lines(p):
    return p.read_text(encoding='utf-8', errors='ignore').splitlines() if p.exists() else []

orig_ref = ORIG_EVAL / 'test.reference.txt.1'
exact_ref = FINAL_EXACT_EVAL / 'test.reference.txt.1'
clean_ref = FINAL_CLEAN_EVAL / 'test.reference.txt.1'

for label, p in [('original', orig_ref), ('final_exact', exact_ref), ('final_clean', clean_ref)]:
    lines = read_lines(p)
    entity_lines = [x for x in lines if '&#x27;' in x or '&amp;#x27;' in x or '&apos;' in x]
    print(label, p, 'exists=', p.exists(), 'lines=', len(lines), 'entity_lines=', len(entity_lines))
    if entity_lines[:3]:
        print(' examples:', entity_lines[:3])

if orig_ref.exists() and exact_ref.exists():
    o = read_lines(orig_ref)
    e = read_lines(exact_ref)
    same_before = sum(a == b for a, b in zip(o, e))
    same_after_unescape = sum(a == html.unescape(html.unescape(b)) for a, b in zip(o, e))
    print('same before:', same_before, '/', min(len(o), len(e)))
    print('same after double html.unescape(generated):', same_after_unescape, '/', min(len(o), len(e)))
````

## 2. Regenerate SGM files safely

Important: `html.escape(..., quote=False)` keeps apostrophes and quotes unchanged. We also run `html.unescape()` first so old `&#x27;` text is converted back to `'`.

````python
SGM_DIR.mkdir(parents=True, exist_ok=True)

def xml_escape_keep_apostrophe(text):
    # normalize any existing entity strings first, then escape only &, <, >
    text = html.unescape(html.unescape(text))
    return html.escape(text, quote=False)

def write_sgm_clean(input_file, output_file, lang, mode):
    lines = input_file.read_text(encoding='utf-8').splitlines()
    with output_file.open('w', encoding='utf-8') as f:
        if mode == 'src':
            f.write('<srcset setid="Myanmar_G2P_v2" srclang="any">\n')
            f.write(f'<doc docid="exp_1_normalized" genre="8000" origlang="{lang}">\n')
        elif mode == 'ref':
            f.write(f'<refset trglang="{lang}" setid="Myanmar_G2P_v2" srclang="any">\n')
            f.write('<doc sysid="ref" docid="exp_1_normalized" genre="8000" origlang="any">\n')
        else:
            raise ValueError(mode)

        for i, line in enumerate(lines, start=1):
            f.write(f'<seg id="{i}">{xml_escape_keep_apostrophe(line)} </seg>\n')

        if mode == 'src':
            f.write('</doc>\n</srcset>\n')
        else:
            f.write('</doc>\n</refset>\n')

for lang in ['my', 'ph']:
    write_sgm_clean(G2P_DIR/f'test_clean.{lang}', SGM_DIR/f'test.{lang}.src.sgm', lang, 'src')
    write_sgm_clean(G2P_DIR/f'test_clean.{lang}', SGM_DIR/f'test.{lang}.ref.sgm', lang, 'ref')

for p in sorted(SGM_DIR.glob('test.*.sgm')):
    text = p.read_text(encoding='utf-8', errors='ignore')
    print(p.name, 'lines=', len(text.splitlines()), 'contains_#x27=', '&#x27;' in text, 'contains_amp_hash=', '&amp;#x27;' in text)
````

## 3. Rerun evaluation cleanly

Option A removes only the evaluation folder and reruns Moses EMS. Training/model artifacts should be reused, but evaluation is rebuilt from fixed SGM.

Run only the experiment you want first.

````python
def sh(cmd, cwd=None, check=True):
    print(f"$ {cmd}")
    return subprocess.run(cmd, shell=True, cwd=str(cwd) if cwd else None, text=True, check=check)

def rerun_ems_evaluation(exp_dir):
    cfgs = list(exp_dir.glob('config.baseline.*'))
    assert cfgs, f'No config found under {exp_dir}'
    cfg = cfgs[0]
    eval_dir = exp_dir / 'evaluation'
    if eval_dir.exists():
        print('Removing old evaluation:', eval_dir)
        shutil.rmtree(eval_dir)
    cmd = f"perl {EXPERIMENT_PERL} -config {cfg} -exec -no-graph 2>&1 | tee {exp_dir/'rerun_after_sgm_fix.log'}"
    sh(cmd, cwd=exp_dir, check=False)

# Recommended first:
# rerun_ems_evaluation(FINAL_EXACT)

# If needed for old clean experiment:
# rerun_ems_evaluation(FINAL_CLEAN)
````

## 4. Manual mteval verification without rerunning EMS

If you already have `test.detokenized.sgm.1`, this directly scores it against the newly fixed SGM reference.

````python
def manual_mteval(eval_dir, src_lang='my', ref_lang='ph', out_name='manual_fixed_sgm_mteval.txt'):
    tst = eval_dir / 'test.detokenized.sgm.1'
    src = SGM_DIR / f'test.{src_lang}.src.sgm'
    ref = SGM_DIR / f'test.{ref_lang}.ref.sgm'
    assert tst.exists(), tst
    assert src.exists(), src
    assert ref.exists(), ref
    out = eval_dir / out_name
    cmd = f"perl {MTEVAL} -c -s {src} -r {ref} -t {tst} 2>&1 | tee {out}"
    sh(cmd, cwd=eval_dir, check=False)
    return out

# Run after SGM regeneration; this does not retrain or decode.
# out = manual_mteval(FINAL_EXACT_EVAL)
# print(out.read_text(encoding='utf-8', errors='ignore')[:3000])
````

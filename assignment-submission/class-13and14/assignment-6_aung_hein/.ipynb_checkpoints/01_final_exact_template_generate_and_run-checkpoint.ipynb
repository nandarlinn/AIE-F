# 01 Final Exact Template Generate and Run

Final exact-template PBSMT notebook.

## Purpose

This is now the safest final setup: it uses your **original high-BLEU `config.baseline` file as the template**.

Instead of using a compact rewritten config, this notebook:

1. Reads `pbsmt-big-normalize/config.baseline`.
2. Prepends only the `[GENERAL]` section per direction.
3. For 3-gram/7-gram variants, changes only active `order = ...` and `max-phrase-length = ...` lines.
4. Leaves every other original Moses EMS setting untouched.

This should reproduce your original 5-gram behavior.

```python
from pathlib import Path
import subprocess, re, shutil, difflib, os

BASE_DIR = Path('/media/kyalkalay/Data/AI-Class/assignment-6_aung_hein')
G2P_DIR = BASE_DIR / 'g2p-par'
MOSES_SRC_DIR = Path('/home/kyalkalay/tools/mosesdecoder')
MOSES_SCRIPT_DIR = MOSES_SRC_DIR / 'scripts'
EXPERIMENT_PERL = MOSES_SCRIPT_DIR / 'ems' / 'experiment.perl'

# IMPORTANT: use the original high-BLEU template exactly.
ORIGINAL_EXP_DIR = BASE_DIR / 'pbsmt-big-normalize'
ORIGINAL_TEMPLATE = ORIGINAL_EXP_DIR / 'config.baseline'
ORIGINAL_BASELINE_DIR = ORIGINAL_EXP_DIR / 'baseline'

FINAL_EXACT_DIR = BASE_DIR / 'pbsmt_final_exact_template_experiments'

print('BASE_DIR:', BASE_DIR)
print('Original template:', ORIGINAL_TEMPLATE, ORIGINAL_TEMPLATE.exists())
print('Original baseline:', ORIGINAL_BASELINE_DIR, ORIGINAL_BASELINE_DIR.exists())
print('Final exact dir:', FINAL_EXACT_DIR)
```

```python
def sh(cmd, cwd=None, check=True):
    print(f"$ {cmd}")
    return subprocess.run(cmd, shell=True, cwd=str(cwd) if cwd else None, text=True, check=check)


def patch_template(template_text, order, max_phrase_length, lm_prune=None):
    """Patch only active experiment values in the original template."""
    text = template_text

    # Patch active LM order line after [LM]. Avoid commented examples.
    text, n_order = re.subn(r'(?m)^order\s*=\s*\d+\s*$', f'order = {order}', text, count=1)
    if n_order != 1:
        raise RuntimeError(f'Could not patch exactly one active order line; patched={n_order}')

    # Patch active phrase length.
    text, n_phrase = re.subn(r'(?m)^max-phrase-length\s*=\s*\d+\s*$', f'max-phrase-length = {max_phrase_length}', text, count=1)
    if n_phrase != 1:
        raise RuntimeError(f'Could not patch exactly one max-phrase-length line; patched={n_phrase}')

    # Optional LM prune patch. Keeps original if lm_prune is None.
    if lm_prune is not None:
        text, n_prune = re.subn(
            r'(?m)^settings\s*=\s*"--prune\s+\'[^\']+\'(.*?)"\s*$',
            lambda m: f'settings = "--prune \'{lm_prune}\'{m.group(1)}"',
            text,
            count=1,
        )
        if n_prune != 1:
            raise RuntimeError(f'Could not patch exactly one LM prune line; patched={n_prune}')
    return text


def write_configs_from_original_template(variants, directions=(('my','ph'), ('ph','my')), clean_existing=False):
    assert ORIGINAL_TEMPLATE.exists(), f'Missing original template: {ORIGINAL_TEMPLATE}'
    template = ORIGINAL_TEMPLATE.read_text(encoding='utf-8', errors='ignore')
    if clean_existing and FINAL_EXACT_DIR.exists():
        shutil.rmtree(FINAL_EXACT_DIR)
    FINAL_EXACT_DIR.mkdir(parents=True, exist_ok=True)

    configs = []
    for v in variants:
        body = patch_template(template, order=v['order'], max_phrase_length=v['max_phrase_length'], lm_prune=v.get('lm_prune'))
        for src, trg in directions:
            pair = f'{src}-{trg}'
            work = FINAL_EXACT_DIR / v['name'] / pair
            work.mkdir(parents=True, exist_ok=True)
            general = f"""# Config file for {pair} ({v['name']})

[GENERAL]
working-dir = {work}
input-extension = {src}
output-extension = {trg}
pair-extension = {pair}

"""
            cfg = general + body
            cfg_path = work / f'config.baseline.{pair}'
            cfg_path.write_text(cfg, encoding='utf-8')
            configs.append(cfg_path)
    return configs


def run_configs(configs, no_graph=True):
    for cfg in configs:
        log_path = cfg.parent / 'run.log'
        cmd = f"perl {EXPERIMENT_PERL} -config {cfg} -exec {'-no-graph' if no_graph else ''} 2>&1 | tee {log_path}"
        sh(cmd, cwd=cfg.parent, check=False)
```

```python
EXACT_VARIANTS = [
    {'name': '5g_original_template_MAIN', 'order': 5, 'max_phrase_length': 9},
    {'name': '3g_original_template_phrase5', 'order': 3, 'max_phrase_length': 5},
    {'name': '3g_original_template_phrase7', 'order': 3, 'max_phrase_length': 7},
    {'name': '7g_original_template_phrase9', 'order': 7, 'max_phrase_length': 9},
    {'name': '7g_original_template_phrase11', 'order': 7, 'max_phrase_length': 11},
]

configs = write_configs_from_original_template(EXACT_VARIANTS, clean_existing=False)
for c in configs:
    print(c)
```

## Sanity check: final main config vs original config

Only `[GENERAL] working-dir` and comments should differ for the main 5g template.

```python
orig = ORIGINAL_BASELINE_DIR / 'my-ph' / 'config.baseline.my-ph'
new = FINAL_EXACT_DIR / '5g_original_template_MAIN' / 'my-ph' / 'config.baseline.my-ph'

def normalize_for_diff(lines):
    out = []
    for x in lines:
        x = re.sub(r'working-dir\s*=.*', 'working-dir = <WORKDIR>', x)
        x = re.sub(r'^# Config file.*$', '# Config file <NORMALIZED>', x)
        out.append(x)
    return out

if orig.exists():
    diff = list(difflib.unified_diff(
        normalize_for_diff(orig.read_text(encoding='utf-8', errors='ignore').splitlines()),
        normalize_for_diff(new.read_text(encoding='utf-8').splitlines()),
        fromfile='original', tofile='exact-template', lineterm=''
    ))
    print('\n'.join(diff[:200]))
    print('Total diff lines:', len(diff))
else:
    print('Original generated config not found yet; skipping diff:', orig)

text = new.read_text(encoding='utf-8')
assert 'decoder-settings = ""' in text
assert 'decoder-settings = "-search-algorithm 1 -cube-pruning-pop-limit 5000 -s 5000"' in text
print('Sanity passed.')
```

## Generate SGM files
The Moses evaluator expects SGM/XML-like source and reference files.

```python
import html

SGM_DIR = G2P_DIR / 'sgm'
SGM_DIR.mkdir(parents=True, exist_ok=True)

def to_sgm(input_file, output_file, lang, mode):
    assert mode in {'src', 'ref'}
    lines = input_file.read_text(encoding='utf-8').splitlines()
    with output_file.open('w', encoding='utf-8') as f:
        if mode == 'src':
            f.write('<srcset setid="Myanmar_G2P_v2" srclang="any">\n')
            f.write(f'<doc docid="exp_1_normalized" genre="8000" origlang="{lang}">\n')
        else:
            f.write(f'<refset trglang="{lang}" setid="Myanmar_G2P_v2" srclang="any">\n')
            f.write('<doc sysid="ref" docid="exp_1_normalized" genre="8000" origlang="any">\n')
        for i, line in enumerate(lines, start=1):
            f.write(f'<seg id="{i}">{html.escape(line)} </seg>\n')
        if mode == 'src':
            f.write('</doc>\n</srcset>\n')
        else:
            f.write('</doc>\n</refset>\n')

for lang in ['my', 'ph']:
    to_sgm(G2P_DIR/f'test_clean.{lang}', SGM_DIR/f'test.{lang}.src.sgm', lang, 'src')
    to_sgm(G2P_DIR/f'test_clean.{lang}', SGM_DIR/f'test.{lang}.ref.sgm', lang, 'ref')

for p in sorted(SGM_DIR.glob('test.*.sgm')):
    print(p, 'lines=', sum(1 for _ in p.open(encoding='utf-8')))
```

## Alignment sanity check

```python
def check_parallel(a, b):
    la = a.read_text(encoding='utf-8').splitlines()
    lb = b.read_text(encoding='utf-8').splitlines()
    print(a.name, b.name, 'line_count_match=', len(la)==len(lb), len(la), len(lb))
    bad = []
    for i, (x, y) in enumerate(zip(la, lb), start=1):
        if len(x.split()) != len(y.split()):
            bad.append((i, x, y, len(x.split()), len(y.split())))
            if len(bad) >= 10:
                break
    print('token-count mismatches, first 10:', len(bad))
    for row in bad[:10]:
        print(row)

check_parallel(G2P_DIR/'train_clean.my', G2P_DIR/'train_clean.ph')
check_parallel(G2P_DIR/'dev_clean.my', G2P_DIR/'dev_clean.ph')
check_parallel(G2P_DIR/'test_clean.my', G2P_DIR/'test_clean.ph')
```

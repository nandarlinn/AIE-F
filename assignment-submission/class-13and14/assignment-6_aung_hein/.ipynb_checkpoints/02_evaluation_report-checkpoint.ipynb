# 02 Evaluation Report

Clean end-to-end PBSMT experiment notebook based on the original assignment setup.

## Purpose

This notebook collects Moses EMS reports and exact-match error counts, then rewrites `README.md` as a real experiment report.

````python
from pathlib import Path
import subprocess, re, shutil, os, difflib

BASE_DIR = Path('/media/kyalkalay/Data/AI-Class/assignment-6_aung_hein')
G2P_DIR = BASE_DIR / 'g2p-par'
SGM_DIR = G2P_DIR / 'sgm'
MOSES_SRC_DIR = Path('/home/kyalkalay/tools/mosesdecoder')
MOSES_SCRIPT_DIR = MOSES_SRC_DIR / 'scripts'
MOSES_BIN_DIR = MOSES_SRC_DIR / 'bin'
EXPERIMENT_PERL = MOSES_SCRIPT_DIR / 'ems' / 'experiment.perl'
MTEVAL = MOSES_SCRIPT_DIR / 'generic' / 'mteval-v13a.pl'

# The original assignment experiment directory.
ORIGINAL_EXP_DIR = BASE_DIR / 'pbsmt-big-normalize'
ORIGINAL_TEMPLATE = ORIGINAL_EXP_DIR / 'config.baseline'
ORIGINAL_BASELINE_DIR = ORIGINAL_EXP_DIR / 'baseline'

# New clean experiment directory.
EXP_DIR = BASE_DIR / 'pbsmt_original_clean_experiments'

print('BASE_DIR:', BASE_DIR)
print('G2P_DIR:', G2P_DIR)
print('SGM_DIR:', SGM_DIR)
print('Original template:', ORIGINAL_TEMPLATE, ORIGINAL_TEMPLATE.exists())
print('Clean experiment dir:', EXP_DIR)
````

````python
import pandas as pd
from datetime import datetime

def read_text(p):
    return p.read_text(encoding='utf-8', errors='ignore') if p.exists() else ''

def clean_out(line):
    return ' '.join(re.sub(r'\|\d+-\d+\|', '', line).split()).strip()

def parse_report(txt):
    m = re.search(r'test:\s*([0-9.]+)\s*\(([0-9.]+)\)\s*BLEU-c\s*;\s*([0-9.]+)\s*\(([0-9.]+)\)\s*BLEU', txt)
    if m:
        return {'bleu_c': float(m.group(1)), 'nist_c': float(m.group(2)), 'bleu': float(m.group(3)), 'nist': float(m.group(4))}
    return {}

def exact_stats(eval_dir):
    f_out = eval_dir / 'test.output.1'
    f_ref = eval_dir / 'test.reference.txt.1'
    if not (f_out.exists() and f_ref.exists()):
        return {}
    total = errors = 0
    for out, ref in zip(f_out.open(encoding='utf-8'), f_ref.open(encoding='utf-8')):
        total += 1
        if clean_out(out) != ' '.join(ref.split()).strip():
            errors += 1
    return {'test_items': total, 'exact_match_errors': errors, 'exact_match_accuracy': round((total-errors)/total*100, 4) if total else None}

rows = []
for eval_dir in sorted(EXP_DIR.glob('*/*/evaluation')):
    row = {
        'experiment': eval_dir.parts[-3],
        'direction': eval_dir.parts[-2],
        'evaluation_dir': str(eval_dir),
        'report_exists': (eval_dir/'report.1').exists(),
    }
    row.update(parse_report(read_text(eval_dir/'report.1')))
    row.update(exact_stats(eval_dir))
    rows.append(row)

results_df = pd.DataFrame(rows)
results_df.to_csv('pbsmt_experiment_results.csv', index=False)
if not results_df.empty:
    display(results_df.sort_values(['direction', 'exact_match_errors'] if 'exact_match_errors' in results_df.columns else ['direction']))
else:
    print('No completed experiment outputs found yet.')
````

````python
result_table = '_No completed experiment outputs were found at report-generation time._'
best_text = '_Pending completed experiment runs._'
if not results_df.empty and 'exact_match_errors' in results_df.columns:
    show_cols = [c for c in ['experiment','direction','bleu_c','nist_c','bleu','nist','test_items','exact_match_errors','exact_match_accuracy'] if c in results_df.columns]
    show = results_df[show_cols].sort_values(['direction','exact_match_errors'])
    result_table = show.to_markdown(index=False)
    best = show.groupby('direction').head(1)
    best_text = '\n'.join([f"- **{r.direction}**: `{r.experiment}`, BLEU-c={r.bleu_c}, exact-match errors={r.exact_match_errors}/{r.test_items}" for _, r in best.iterrows()])

readme = f"""# Myanmar G2P PBSMT Experiment Report

## Abstract

This experiment evaluates phrase-based statistical machine translation for Myanmar grapheme-to-phoneme conversion. The main system is a 5-gram PBSMT model trained with Moses EMS. Additional 3-gram and 7-gram systems are included to compare lower and higher phoneme-context settings.

## Data

The experiment uses the myG2P Myanmar grapheme-to-phoneme dictionary data. The source side is Myanmar grapheme syllables (`.my`), and the target side is phoneme representation (`.ph`). The cleaned split files are:

```text
g2p-par/train_clean.my
g2p-par/train_clean.ph
g2p-par/dev_clean.my
g2p-par/dev_clean.ph
g2p-par/test_clean.my
g2p-par/test_clean.ph
```

## Normalization

The dataset was normalized with `syl-Normalizer` v0.6 and `selected_syl_dictionary_13Feb2024.sorted.txt`. The cleaned test split was converted to Moses SGM format using original-style segment text handling so that phoneme apostrophes such as `te'` remain unchanged.

## Experimental Setup

The configs are generated from the original assignment template:

```text
pbsmt-big-normalize/config.baseline
```

The main system is:

```text
pbsmt_5gram_main
```

Comparison systems:

```text
pbsmt_3gram_phrase5
pbsmt_3gram_phrase7
pbsmt_7gram_phrase9
pbsmt_7gram_phrase11
```

The original observed `my-ph` mteval score from the assignment run was BLEU `0.7587` and NIST `10.6107`, equivalent to approximately `75.87` BLEU when reported as a percentage.

## Results

Generated: {datetime.now().strftime('%Y-%m-%d %H:%M')}

{result_table}

## Comparison

{best_text}

## Findings

- The main 5-gram setup follows the original Moses EMS template rather than a rewritten compact config.
- SGM/reference text handling is important for this task because phoneme apostrophes are part of the target tokens.
- Exact-match error count is more interpretable than BLEU alone for G2P because each output line should match a specific pronunciation sequence.
- The 3-gram systems test whether lower context reduces sparsity.
- The 7-gram systems test whether longer phoneme context helps ambiguous syllables, but they may be more sensitive to data sparsity.

## Conclusion

The 5-gram PBSMT system generated from the original template should be treated as the main experiment. The 3-gram and 7-gram systems are comparison runs. Model selection should prioritize exact-match error count, then BLEU/NIST, then root-cause error analysis.

## Sources

- Ye Kyaw Thu, `myG2P`: https://github.com/ye-kyaw-thu/myG2P
- Ye Kyaw Thu, `syl-Normalizer`: https://github.com/ye-kyaw-thu/syl-Normalizer
"""
Path('README.md').write_text(readme, encoding='utf-8')
print('README.md updated')
````

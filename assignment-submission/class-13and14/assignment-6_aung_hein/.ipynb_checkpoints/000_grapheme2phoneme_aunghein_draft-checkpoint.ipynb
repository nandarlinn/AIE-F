## SGM
Let change data to SGM.

```python
!mkdir g2p-par/sgm/
```

```python
%%writefile g2p-par/sgm/generate_sgms.pl
#!/usr/bin/perl
use strict;

# written by Ye, NECTEC
# for MTRSS, YTU, Myanmar

my @langs;

#foreach my $trainFile ( </home/ros/experiment/my-rk/data/train.[a-z][a-z]> )
foreach my $trainFile ( <../test_clean.[a-z][a-z]> )
{                        
        $trainFile =~ m/test_clean.([a-z][a-z])/;
        push @langs, $1;
}

foreach my $lang (@langs)
{ 
    `perl ./ref2sgm.pl $lang > test.$lang.ref.sgm`;
    `perl ./src2sgm.pl $lang > test.$lang.src.sgm`;
}
```

```python
%%writefile g2p-par/sgm/ref2sgm.pl

#!/usr/bin/perl
use strict;

# written by Ye, NECTEC
# for MTRSS, YTU, Myanmar

my $trg = shift;

#print "<refset trglang=\"$trg\" setid=\"sl_data\" srclang=\"any\">\n";
print "<refset trglang=\"$trg\" setid=\"Myanmar_G2P_v2\" srclang=\"any\">\n";
print "<doc sysid=\"ref\" docid=\"exp_1_normalized\" genre=\"8000\" origlang=\"any\">\n";

#open FILE, "/home/ros/experiment/my-rk/data/test.$trg" or die;
open FILE, "../test_clean.$trg" or die;
             
my $id=1;

while( <FILE> )
{
	chomp;
	
	print "<seg id=\"$id\">$_ </seg>\n";
	$id++;
}

print "</doc>\n</refset>\n";
```

```python
%%writefile g2p-par/sgm/src2sgm.pl

#!/usr/bin/perl
use strict;

# written by Ye, NECTEC
# for MTRSS, YTU, Myanmar

my $src = shift;

print "<srcset setid=\"Myanmar_G2P_v2\" srclang=\"any\">\n";
print "<doc docid=\"exp_1_normalized\" genre=\"8000\" origlang=\"$src\">\n";

#open FILE, "/home/ros/experiment/my-rk/data/myph.$src" or die;
open FILE, "../test_clean.$src" or die;

my $id=1;

while( <FILE> )
{
	chomp;
	
	print "<seg id=\"$id\">$_ </seg>\n";
	$id++;
}

print "</doc>\n</srcset>\n";
```

```python
import re

# File paths based on your setup
file_in = "pbsmt-big-normalize/baseline/my-ph/evaluation/test.input.txt.1"
file_out = "pbsmt-big-normalize/baseline/my-ph/evaluation/test.output.1"
file_ref = "pbsmt-big-normalize/baseline/my-ph/evaluation/test.reference.txt.1"

print(f"{'Source Grapheme':<18} | {'Moses Output (Cleaned)':<25} | {'Perfect Reference':<25}")
print("-" * 80)

error_count = 0

with open(file_in, 'r', encoding='utf-8') as f_in, \
     open(file_out, 'r', encoding='utf-8') as f_out, \
     open(file_ref, 'r', encoding='utf-8') as f_ref:
    
    for src, out, ref in zip(f_in, f_out, f_ref):
        src_clean = src.strip()
        ref_clean = " ".join(ref.split()).strip()
        
        # 1. Strip out the Moses alignment tags (e.g., |0-0|, |2-3|)
        out_cleaned = re.sub(r'\|\d+-\d+\|', '', out)
        # 2. Collapse multiple spaces down to clean single tokens
        out_cleaned = " ".join(out_cleaned.split()).strip()
        
        # 3. If they don't match, it's a translation error!
        if out_cleaned != ref_clean:
            print(f"{src_clean:<18} | {out_cleaned:<25} | {ref_clean:<25}")
            error_count += 1

print("-" * 80)
print(f"Found {error_count} translation errors to analyze.")
```

```python
import re
import pandas as pd
from collections import defaultdict, Counter

# --- CONFIGURATION PATHS ---
file_train_src = "g2p-par/train_clean.my"
file_train_trg = "g2p-par/train_clean.ph"
file_test_src  = "pbsmt-big-normalize/baseline/my-ph/evaluation/test.input.txt.1"
file_test_out  = "pbsmt-big-normalize/baseline/my-ph/evaluation/test.output.1"
file_test_ref  = "pbsmt-big-normalize/baseline/my-ph/evaluation/test.reference.txt.1"

# --- STEP 1: PARSE TRAINING DATA DISTRIBUTION ---
print("Indexing training data distributions...")
train_counts = defaultdict(Counter)

with open(file_train_src, 'r', encoding='utf-8') as f_src, \
     open(file_train_trg, 'r', encoding='utf-8') as f_trg:
    for src_line, trg_line in zip(f_src, f_trg):
        s_tokens = src_line.strip().split()
        t_tokens = trg_line.strip().split()
        # Process strictly aligned token streams
        if len(s_tokens) == len(t_tokens):
            for s, t in zip(s_tokens, t_tokens):
                train_counts[s][t] += 1

# --- STEP 2: PARSE TEST MAPPED ERRORS ---
print("Analyzing test translation alignments...")
test_raw_errors = []

with open(file_test_src, 'r', encoding='utf-8') as f_in, \
     open(file_test_out, 'r', encoding='utf-8') as f_out, \
     open(file_test_ref, 'r', encoding='utf-8') as f_ref:
    
    for src, out, ref in zip(f_in, f_out, f_ref):
        src_tokens = src.strip().split()
        ref_tokens = ref.strip().split()
        
        if len(src_tokens) != len(ref_tokens):
            continue  # Fallback guardrail
            
        moses_phrases = re.findall(r'(.+?)\s*\|(\d+)-(\d+)\|', out)
        covered_indices = set()
        
        for pred_phrase, start_str, end_str in moses_phrases:
            start, end = int(start_str), int(end_str)
            pred_phrase = " ".join(pred_phrase.split()).strip()
            
            if end < len(ref_tokens):
                src_chunk = " ".join(src_tokens[start:end+1])
                ref_chunk = " ".join(ref_tokens[start:end+1])
                for i in range(start, end+1):
                    covered_indices.add(i)
                if pred_phrase != ref_chunk:
                    test_raw_errors.append((src_chunk, pred_phrase, ref_chunk))
                    
        # Catch explicit deletions
        for i in range(len(src_tokens)):
            if i not in covered_indices:
                test_raw_errors.append((src_tokens[i], "[DELETED]", ref_tokens[i]))

# --- STEP 3: CONSTRUCT PANDAS DATAFRAME & CALCULATE METRICS ---
error_records = []
for src_chunk, pred, ref in test_raw_errors:
    # Get training baseline context
    tr_counts = train_counts.get(src_chunk, Counter())
    total_train_instances = sum(tr_counts.values())
    
    train_pred_freq = tr_counts.get(pred, 0)
    train_ref_freq = tr_counts.get(ref, 0)
    
    # Calculate conditional probabilities observed in training data
    p_pred_given_src = (train_pred_freq / total_train_instances) if total_train_instances > 0 else 0.0
    p_ref_given_src  = (train_ref_freq / total_train_instances) if total_train_instances > 0 else 0.0
    
    # Categorize the root cause of the error based on training behavior
    if total_train_instances == 0:
        error_cause = "Out of Vocabulary (OOV)"
    elif train_pred_freq > 0 and train_ref_freq > 0:
        error_cause = "Train Data Ambiguity (Both Exist)"
    elif train_ref_freq == 0 and train_pred_freq > 0:
        error_cause = "Reference Missing from Training Token"
    elif train_pred_freq == 0 and train_ref_freq > 0:
        error_cause = "Decoder Hallucination / Over-generalization"
    else:
        error_cause = "Unclassified Structural Swap"
        
    error_records.append({
        "Source_Syllable": src_chunk,
        "Moses_Output": pred,
        "True_Reference": ref,
        "Train_Total_Seen": total_train_instances,
        "Train_Prob_Pred": round(p_pred_given_src, 3),
        "Train_Prob_Ref": round(p_ref_given_src, 3),
        "Root_Cause": error_cause
    })
df_errors = pd.DataFrame(error_records)

# Group duplicates to show the most frequent failure pairs
df_leaderboard = df_errors.groupby(["Source_Syllable", "Moses_Output", "True_Reference", "Root_Cause", "Train_Prob_Pred", "Train_Prob_Ref"]).size().reset_index(name="Error_Count")
df_leaderboard = df_leaderboard.sort_values(by="Error_Count", ascending=False).reset_index(drop=True)

# --- STEP 4: PRINT BREAKDOWN STATS ---
print("\n=== SYSTEMIC ROOT CAUSE ANALYSIS ===")
stats = df_errors["Root_Cause"].value_counts(normalize=True) * 100
counts = df_errors["Root_Cause"].value_counts()
stats_df = pd.DataFrame({"Total Errors": counts, "Percentage": stats.round(2).astype(str) + "%"})
stats_df
```

```python
%%writefile pbsmt_3gram/generate_config.pl
#!/usr/bin/perl

# Preparing configuration files for PBSMT experiments
# Output: one configuration file for baseline or PBSMT
# *** Before you run this script, you should prepare training, development and test parallel-data in advance!!!

# Written by Finch-san and Ye
# We used this script for running our NAACL paper:
# Ye Kyaw Thu, Andrew Finch and Eiichiro Sumita, 
# "Interlocking Phrases in Phrase-based Statistical Machine Translation",
# In Proceedings of the NAACL-HLT 2016, June 12-17, San Diego, US, pp. 1076-1081.

# Last updated 9 Oct 2019, @Machine Translation Research Summer School, YTU, Myanmar

use strict;

my @langs;

# you have to update the PBSMT experiment path!!!
#my $smtpath = "/home/lar/experiment/kachin-myanmar/demo-mykc-smt";
my $smtpath = "/media/kyalkalay/Data/AI-Class/assignment-6_aung_hein/pbsmt_3gram";

# you have to update the data path for running PBSMT experiment!!!

#foreach my $trainFile ( </panfs/panltg2/users/finch/expt/multibtec/corpus/train.[a-z][a-z]> )
#foreach my $trainFile ( </panfs/panmt/users/ye/cluster/corpus/train.[a-z][a-z]>)
#foreach my $trainFile ( </home/tha/experiment/my-rk/data/train.[a-z][a-z]>)
#foreach my $trainFile ( </home/lar/experiment/kachin-myanmar/demo-mykc-smt/4demo/train.[a-z][a-z]>)
foreach my $trainFile ( </media/kyalkalay/Data/AI-Class/assignment-6_aung_hein/g2p-par/train_clean.[a-z][a-z]>)
{                        
        $trainFile =~ m/train_clean.([a-z][a-z])/;
        
        push @langs, $1;
}

foreach my $expt (qw/baseline/)
{

die("$smtpath/$expt") if (-d "$smtpath/$expt");
`mkdir $smtpath/$expt`;
    foreach my $src (@langs)
    {
    	foreach my $trg (@langs)
    	{
    		next if $src eq $trg;
            #next if $src ne "my" and $trg ne "my";
    
    		my $pair = "${src}-${trg}";
    		my $configFile = "$smtpath/$expt/$pair/config.${expt}.$pair";

    		`mkdir $smtpath/$expt/$pair`;

    		open FILE, ">$configFile" or die;
    
    		print FILE "# Config file for $pair ($expt)\n";
    		print FILE "\n[GENERAL]\n";
    		print FILE "working-dir = $smtpath/$expt/$pair\n";
    		print FILE "input-extension = ${src}\n";
    		print FILE "output-extension = ${trg}\n";
    		print FILE "pair-extension = ${pair}\n";
    		print FILE "\n";
    		close FILE;
     	
     	    `cat $configFile config.$expt > tmp`;
    	    `mv tmp $configFile`;
   	    }
    }
}
```

#### Error Checking
let check the error and confusion.

```python
import re

# File paths based on your setup
file_in = "pbsmt_3gram/baseline/my-ph/evaluation/test.input.txt.1"
file_out = "pbsmt_3gram/baseline/my-ph/evaluation/test.output.1"
file_ref = "pbsmt_3gram/baseline/my-ph/evaluation/test.reference.txt.1"

print(f"{'Source Grapheme':<18} | {'Moses Output (Cleaned)':<25} | {'Perfect Reference':<25}")
print("-" * 80)

error_count = 0

with open(file_in, 'r', encoding='utf-8') as f_in, \
     open(file_out, 'r', encoding='utf-8') as f_out, \
     open(file_ref, 'r', encoding='utf-8') as f_ref:
    
    for src, out, ref in zip(f_in, f_out, f_ref):
        src_clean = src.strip()
        ref_clean = " ".join(ref.split()).strip()
        
        # 1. Strip out the Moses alignment tags (e.g., |0-0|, |2-3|)
        out_cleaned = re.sub(r'\|\d+-\d+\|', '', out)
        # 2. Collapse multiple spaces down to clean single tokens
        out_cleaned = " ".join(out_cleaned.split()).strip()
        
        # 3. If they don't match, it's a translation error!
        if out_cleaned != ref_clean:
            print(f"{src_clean:<18} | {out_cleaned:<25} | {ref_clean:<25}")
            error_count += 1

print("-" * 80)
print(f"Found {error_count} translation errors to analyze.")
```

```python
import re

# File paths based on your setup
file_in = "pbsmt_7gram/baseline/my-ph/evaluation/test.input.txt.1"
file_out = "pbsmt_7gram/baseline/my-ph/evaluation/test.output.1"
file_ref = "pbsmt_7gram/baseline/my-ph/evaluation/test.reference.txt.1"

print(f"{'Source Grapheme':<18} | {'Moses Output (Cleaned)':<25} | {'Perfect Reference':<25}")
print("-" * 80)

error_count = 0

with open(file_in, 'r', encoding='utf-8') as f_in, \
     open(file_out, 'r', encoding='utf-8') as f_out, \
     open(file_ref, 'r', encoding='utf-8') as f_ref:
    
    for src, out, ref in zip(f_in, f_out, f_ref):
        src_clean = src.strip()
        ref_clean = " ".join(ref.split()).strip()
        
        # 1. Strip out the Moses alignment tags (e.g., |0-0|, |2-3|)
        out_cleaned = re.sub(r'\|\d+-\d+\|', '', out)
        # 2. Collapse multiple spaces down to clean single tokens
        out_cleaned = " ".join(out_cleaned.split()).strip()
        
        # 3. If they don't match, it's a translation error!
        if out_cleaned != ref_clean:
            print(f"{src_clean:<18} | {out_cleaned:<25} | {ref_clean:<25}")
            error_count += 1

print("-" * 80)
print(f"Found {error_count} translation errors to analyze.")
```

```python
import re
import pandas as pd
from collections import defaultdict, Counter

# --- CONFIGURATION PATHS ---
file_train_src = "g2p-par/train_clean.my"
file_train_trg = "g2p-par/train_clean.ph"
file_test_src  = "pbsmt_7gram/baseline/my-ph/evaluation/test.input.txt.1"
file_test_out  = "pbsmt_7gram/baseline/my-ph/evaluation/test.output.1"
file_test_ref  = "pbsmt_7gram/baseline/my-ph/evaluation/test.reference.txt.1"

# --- STEP 1: PARSE TRAINING DATA DISTRIBUTION ---
print("Indexing training data distributions...")
train_counts = defaultdict(Counter)

with open(file_train_src, 'r', encoding='utf-8') as f_src, \
     open(file_train_trg, 'r', encoding='utf-8') as f_trg:
    for src_line, trg_line in zip(f_src, f_trg):
        s_tokens = src_line.strip().split()
        t_tokens = trg_line.strip().split()
        # Process strictly aligned token streams
        if len(s_tokens) == len(t_tokens):
            for s, t in zip(s_tokens, t_tokens):
                train_counts[s][t] += 1

# --- STEP 2: PARSE TEST MAPPED ERRORS ---
print("Analyzing test translation alignments...")
test_raw_errors = []

with open(file_test_src, 'r', encoding='utf-8') as f_in, \
     open(file_test_out, 'r', encoding='utf-8') as f_out, \
     open(file_test_ref, 'r', encoding='utf-8') as f_ref:
    
    for src, out, ref in zip(f_in, f_out, f_ref):
        src_tokens = src.strip().split()
        ref_tokens = ref.strip().split()
        
        if len(src_tokens) != len(ref_tokens):
            continue  # Fallback guardrail
            
        moses_phrases = re.findall(r'(.+?)\s*\|(\d+)-(\d+)\|', out)
        covered_indices = set()
        
        for pred_phrase, start_str, end_str in moses_phrases:
            start, end = int(start_str), int(end_str)
            pred_phrase = " ".join(pred_phrase.split()).strip()
            
            if end < len(ref_tokens):
                src_chunk = " ".join(src_tokens[start:end+1])
                ref_chunk = " ".join(ref_tokens[start:end+1])
                for i in range(start, end+1):
                    covered_indices.add(i)
                if pred_phrase != ref_chunk:
                    test_raw_errors.append((src_chunk, pred_phrase, ref_chunk))
                    
        # Catch explicit deletions
        for i in range(len(src_tokens)):
            if i not in covered_indices:
                test_raw_errors.append((src_tokens[i], "[DELETED]", ref_tokens[i]))

# --- STEP 3: CONSTRUCT PANDAS DATAFRAME & CALCULATE METRICS ---
error_records = []
for src_chunk, pred, ref in test_raw_errors:
    # Get training baseline context
    tr_counts = train_counts.get(src_chunk, Counter())
    total_train_instances = sum(tr_counts.values())
    
    train_pred_freq = tr_counts.get(pred, 0)
    train_ref_freq = tr_counts.get(ref, 0)
    
    # Calculate conditional probabilities observed in training data
    p_pred_given_src = (train_pred_freq / total_train_instances) if total_train_instances > 0 else 0.0
    p_ref_given_src  = (train_ref_freq / total_train_instances) if total_train_instances > 0 else 0.0
    
    # Categorize the root cause of the error based on training behavior
    if total_train_instances == 0:
        error_cause = "Out of Vocabulary (OOV)"
    elif train_pred_freq > 0 and train_ref_freq > 0:
        error_cause = "Train Data Ambiguity (Both Exist)"
    elif train_ref_freq == 0 and train_pred_freq > 0:
        error_cause = "Reference Missing from Training Token"
    elif train_pred_freq == 0 and train_ref_freq > 0:
        error_cause = "Decoder Hallucination / Over-generalization"
    else:
        error_cause = "Unclassified Structural Swap"
        
    error_records.append({
        "Source_Syllable": src_chunk,
        "Moses_Output": pred,
        "True_Reference": ref,
        "Train_Total_Seen": total_train_instances,
        "Train_Prob_Pred": round(p_pred_given_src, 3),
        "Train_Prob_Ref": round(p_ref_given_src, 3),
        "Root_Cause": error_cause
    })

df_errors = pd.DataFrame(error_records)

# Group duplicates to show the most frequent failure pairs
df_leaderboard = df_errors.groupby(["Source_Syllable", "Moses_Output", "True_Reference", "Root_Cause", "Train_Prob_Pred", "Train_Prob_Ref"]).size().reset_index(name="Error_Count")
df_leaderboard = df_leaderboard.sort_values(by="Error_Count", ascending=False).reset_index(drop=True)

# --- STEP 4: PRINT BREAKDOWN STATS ---
print("\n=== SYSTEMIC ROOT CAUSE ANALYSIS ===")
stats = df_errors["Root_Cause"].value_counts(normalize=True) * 100
counts = df_errors["Root_Cause"].value_counts()
stats_df = pd.DataFrame({"Total Errors": counts, "Percentage": stats.round(2).astype(str) + "%"})
stats_df
```

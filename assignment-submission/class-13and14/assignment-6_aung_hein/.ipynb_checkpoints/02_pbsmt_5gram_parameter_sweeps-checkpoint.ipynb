# 02 PBSMT 5-Gram Parameter Sweeps

Generated PBSMT 5-gram experiment notebook.

## Goal
Try several 5-gram PBSMT variants while keeping the LM order fixed at `5`.

Recommended sweep dimensions:

1. `max-phrase-length`: 7, 9, 11
2. decoder search strength: pop-limit/stack 5000 vs 10000
3. LM pruning: `0 0 1` vs `0 0 0`
4. MERT n-best size: 100 vs 200

For G2P, monotonic decoding is usually preferred, so all variants keep `-distortion-limit 0` unless explicitly testing reordering.

```python
from pathlib import Path
import os, re, json, subprocess, textwrap, shutil

# === EDIT THESE PATHS ONLY IF YOUR MACHINE PATHS CHANGE ===
BASE_DIR = Path('/media/kyalkalay/Data/AI-Class/assignment-6_aung_hein')
G2P_DIR = BASE_DIR / 'g2p-par'
MOSES_SRC_DIR = Path('/home/kyalkalay/tools/mosesdecoder')
MOSES_SCRIPT_DIR = MOSES_SRC_DIR / 'scripts'
MOSES_BIN_DIR = MOSES_SRC_DIR / 'bin'
EXPERIMENT_PERL = MOSES_SCRIPT_DIR / 'ems' / 'experiment.perl'
EXTERNAL_BIN_DIR = Path('/usr/local/bin')

# Experiment outputs will be created under BASE_DIR
assert BASE_DIR.exists(), f'BASE_DIR not found: {BASE_DIR}'
print('BASE_DIR:', BASE_DIR)
print('G2P_DIR:', G2P_DIR)
print('Moses:', MOSES_SRC_DIR)
```

```python
def sh(cmd, cwd=None, check=True):
    """Run shell command and stream output."""
    print(f"$ {cmd}")
    return subprocess.run(cmd, shell=True, cwd=str(cwd) if cwd else None, check=check, text=True)

EXPERIMENT_PERL = MOSES_SCRIPT_DIR / 'ems' / 'experiment.perl'

def moses_ems_config(
    working_dir,
    input_ext,
    output_ext,
    order=5,
    max_phrase_length=9,
    lm_prune="0 0 1",
    lm_memory="20%",
    lexicalized_reordering="msd-bidirectional-fe",
    decoder_settings="-search-algorithm 1 -cube-pruning-pop-limit 5000 -s 5000 -distortion-limit 0",
    tuning_decoder_settings="-distortion-limit 0",
    nbest=100,
    cores=8,
):
    """Create a Moses EMS config specialized for Myanmar G2P PBSMT."""
    return f"""
[GENERAL]
working-dir = {working_dir}
input-extension = {input_ext}
output-extension = {output_ext}
pair-extension = {input_ext}-{output_ext}

moses-src-dir = {MOSES_SRC_DIR}
moses-bin-dir = $moses-src-dir/bin
moses-script-dir = $moses-src-dir/scripts
external-bin-dir = {EXTERNAL_BIN_DIR}
myph-data = {G2P_DIR}
decoder = $moses-bin-dir/moses
generic-parallelizer = $moses-script-dir/ems/support/generic-parallelizer.perl
cores = {cores}

[CORPUS]
max-sentence-length = 80

[CORPUS:myph]
raw-stem = $myph-data/train_clean

[LM]
lmplz = $moses-bin-dir/lmplz
lm-training = "$moses-script-dir/ems/support/lmplz-wrapper.perl -bin $moses-bin-dir/lmplz"
settings = "--prune '{lm_prune}' -T $working-dir/lm -S {lm_memory} --discount_fallback"
order = {order}
lm-binarizer = $moses-bin-dir/build_binary
type = 8

[LM:myph]
raw-corpus = $myph-data/train_clean.$output-extension
lm-binarizer = $moses-bin-dir/build_binary
type = 8

[MML] IGNORE

[TRAINING]
script = $mosu-script_dir/training/train-model.perl
""".replace('$mosu-script_dir', '$moses-script-dir') + f"""
parallel = yes
run-giza-in-parts = 5
alignment-symmetrization-method = grow-diag-final-and
lexicalized-reordering = {lexicalized_reordering}
max-phrase-length = {max_phrase_length}
score-settings = "--GoodTuring"

[TUNING]
tuning-script = $moses-script-dir/training/mert-moses.pl
tuning-settings = "-mertdir $moses-bin-dir"
raw-input = $myph-data/dev_clean.$input-extension
raw-reference = $myph-data/dev_clean.$output-extension
nbest = {nbest}
filter-settings = ""
decoder-settings = "{tuning_decoder_settings}"

[EVALUATION]
decoder-settings = "{decoder_settings}"
wrapping-script = "$moses-script-dir/ems/support/wrap-xml.perl $output-extension"
nist-bleu = $moses-script-dir/generic/mteval-v13a.pl
nist-bleu-c = "$mosu_script_dir/generic/mteval-v13a.pl -c"
analysis = $moses-script-dir/ems/support/analysis.perl
analyze-coverage = yes
report-segmentation = yes
report-precision-by-coverage = yes
""".replace('$mosu_script_dir', '$moses-script-dir') + f"""

[EVALUATION:test]
input-sgm = $myph-data/sgm/test.$input-extension.src.sgm
reference-sgm = $myph-data/sgm/test.$output-extension.ref.sgm
wrapping-frame = $input-sgm

[REPORTING]
""".strip() + "\n"


def write_experiment_configs(exp_name, variants, directions=(('my','ph'), ('ph','my'))):
    """Write config.baseline.<src-trg> for each variant and direction."""
    exp_root = BASE_DIR / exp_name
    exp_root.mkdir(parents=True, exist_ok=True)
    written = []
    for variant in variants:
        variant_name = variant['name']
        for src, trg in directions:
            pair = f'{src}-{trg}'
            work = exp_root / variant_name / pair
            work.mkdir(parents=True, exist_ok=True)
            cfg = moses_ems_config(
                working_dir=work,
                input_ext=src,
                output_ext=trg,
                order=5,
                max_phrase_length=variant.get('max_phrase_length', 9),
                lm_prune=variant.get('lm_prune', '0 0 1'),
                lm_memory=variant.get('lm_memory', '20%'),
                lexicalized_reordering=variant.get('lexicalized_reordering', 'msd-bidirectional-fe'),
                decoder_settings=variant.get('decoder_settings', '-search-algorithm 1 -cube-pruning-pop-limit 5000 -s 5000 -distortion-limit 0'),
                tuning_decoder_settings=variant.get('tuning_decoder_settings', '-distortion-limit 0'),
                nbest=variant.get('nbest', 100),
                cores=variant.get('cores', 8),
            )
            cfg_path = work / f'config.baseline.{pair}'
            cfg_path.write_text(cfg, encoding='utf-8')
            written.append(cfg_path)
    print(f'Wrote {len(written)} config files under {exp_root}')
    return written

def run_configs(config_paths, no_graph=True):
    """Run Moses EMS for a list of config files. This can take a long time."""
    for cfg in config_paths:
        log_path = cfg.parent / 'run.log'
        cmd = f"perl {EXPERIMENT_PERL} -config {cfg} -exec {'-no-graph' if no_graph else ''} 2>&1 | tee {log_path}"
        sh(cmd, cwd=cfg.parent, check=False)
```

```python
VARIANTS = [
    {
        'name': 'v01_phrase7_prune001',
        'max_phrase_length': 7,
        'lm_prune': '0 0 1',
        'decoder_settings': '-search-algorithm 1 -cube-pruning-pop-limit 5000 -s 5000 -distortion-limit 0',
        'tuning_decoder_settings': '-distortion-limit 0',
        'nbest': 100,
    },
    {
        'name': 'v02_phrase9_prune001_main',
        'max_phrase_length': 9,
        'lm_prune': '0 0 1',
        'decoder_settings': '-search-algorithm 1 -cube-pruning-pop-limit 5000 -s 5000 -distortion-limit 0',
        'tuning_decoder_settings': '-distortion-limit 0',
        'nbest': 100,
    },
    {
        'name': 'v03_phrase11_prune001',
        'max_phrase_length': 11,
        'lm_prune': '0 0 1',
        'decoder_settings': '-search-algorithm 1 -cube-pruning-pop-limit 5000 -s 5000 -distortion-limit 0',
        'tuning_decoder_settings': '-distortion-limit 0',
        'nbest': 100,
    },
    {
        'name': 'v04_phrase9_prune000_less_pruned',
        'max_phrase_length': 9,
        'lm_prune': '0 0 0',
        'decoder_settings': '-search-algorithm 1 -cube-pruning-pop-limit 5000 -s 5000 -distortion-limit 0',
        'tuning_decoder_settings': '-distortion-limit 0',
        'nbest': 100,
    },
    {
        'name': 'v05_phrase9_stronger_search',
        'max_phrase_length': 9,
        'lm_prune': '0 0 1',
        'decoder_settings': '-search-algorithm 1 -cube-pruning-pop-limit 10000 -s 10000 -distortion-limit 0',
        'tuning_decoder_settings': '-distortion-limit 0',
        'nbest': 100,
    },
    {
        'name': 'v06_phrase9_nbest200',
        'max_phrase_length': 9,
        'lm_prune': '0 0 1',
        'decoder_settings': '-search-algorithm 1 -cube-pruning-pop-limit 5000 -s 5000 -distortion-limit 0',
        'tuning_decoder_settings': '-distortion-limit 0',
        'nbest': 200,
    },
]

configs = write_experiment_configs('pbsmt_5gram_experiments', VARIANTS)
print('Total configs:', len(configs))
```

## Single-Video Inference with Transformer Model

## Batch Inference

**အရင်ဆုံး infer.py ပရိုဂရမ်ရဲ့ commandline option တွေကို --help နဲ့ခေါ်ကြည့်ကြရအောင်။**  

````python
!time python src/infer.py --help
````

အထက်မှာ testing လုပ်ပြခဲ့တဲ့ ဗီဒီယိုဖိုင် ၅ဖိုင်စလုံးကို Bi-LSTM မော်ဒယ်နဲ့ ဘာသာပြန်ခိုင်းပါမယ်။ 
ဗီဒီယိုဖိုင်တွေ သိမ်းထားတဲ့ ဖိုလ်ဒါလမ်းကြောင်းကိုတော့ --batch ဆိုတဲ့ option နဲ့ ပေးရပါမယ်။ အောက်ပါအတိုင်းပါ။  

````python
!time python src/infer.py \
      --checkpoint results/exp_bilstm/checkpoints/best.pth \
      --config     config/config.yaml \
      --batch      data/sample4inference/ \
      --top_k      3
````
